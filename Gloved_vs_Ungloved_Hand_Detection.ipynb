```python
import os
import json
import cv2
from google.colab import drive
from ultralytics import YOLO
from IPython.display import display, Image
```

```python
# Mount Google Drive
drive.mount('/content/drive')
```

```python
INPUT_FOLDER = '/content/drive/MyDrive/Hand Detection/train/images'
OUTPUT_FOLDER = '/content/drive/MyDrive/Hand Detection/output_annotated'
LOG_FOLDER = '/content/drive/MyDrive/Hand Detection/output_logs'
```

```python
from ultralytics import YOLO

# Load a pre-trained model to start training from
model = YOLO('yolov8n.pt')
```

```python
try:
    # Capture the results of the training
    results = model.train(
        data='/content/drive/MyDrive/Hand Detection/data.yaml',
        epochs=50,
        imgsz=640,
        project='/content/drive/MyDrive/Hand Detection/training_runs',
        name='hand_detection_model'
    )
    # Dynamically get the path to the best model from the results object
    TRAINED_MODEL_PATH = os.path.join(results.save_dir, 'weights', 'best.pt')
    print("‚úÖ Training completed successfully!")
    print(f"Your best model is saved at: {TRAINED_MODEL_PATH}")

except Exception as e:
    print(f"‚ùå Error during training: {e}")
    exit()
```

```python
# Load the custom-trained model
try:
    trained_model = YOLO(TRAINED_MODEL_PATH)
    print(f"‚úÖ Successfully loaded custom model from {TRAINED_MODEL_PATH}")
except Exception as e:
    print(f"‚ùå Error loading trained model: {e}")
    exit()
```

```python
# Getting the list of target class names from the trained model itself
# This avoids manual entry and potential typos
TARGET_CLASSES = trained_model.names
TARGET_CLASSES
```

```python
# Confidence threshold for detections
CONFIDENCE_THRESHOLD = 0.5
```

```python
# Create output directories if they don't exist
os.makedirs(OUTPUT_FOLDER, exist_ok=True)
os.makedirs(LOG_FOLDER, exist_ok=True)
```

```python
# Get a list of image files to process
if not os.path.exists(INPUT_FOLDER):
    print(f"‚ùå Input folder '{INPUT_FOLDER}' does not exist.")
else:
    image_files = [f for f in os.listdir(INPUT_FOLDER) if f.lower().endswith('.jpg')]

    if not image_files:
        print(f"‚ÑπÔ∏è No .jpg images found in '{INPUT_FOLDER}'.")
    else:
        print(f"‚û°Ô∏è Found {len(image_files)} images to process...")

        for filename in image_files:
            image_path = os.path.join(INPUT_FOLDER, filename)

            try:
                # Perform detection with the custom model
                results = trained_model(source=image_path, conf=CONFIDENCE_THRESHOLD, save=True, project=OUTPUT_FOLDER, name='runs', exist_ok=True)

                # Process detection results and log to JSON
                detections_list = []
                for r in results:
                    for box in r.boxes:
                        x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)
                        confidence = float(box.conf[0].cpu().numpy())
                        class_id = int(box.cls[0].cpu().numpy())
                        label = trained_model.names[class_id]

                        detections_list.append({
                            "label": label,
                            "confidence": round(confidence, 2),
                            "bbox": [int(x1), int(y1), int(x2), int(y2)]
                        })

                # Save detections to a JSON file
                json_file_path = os.path.join(LOG_FOLDER, f"{os.path.splitext(filename)[0]}.json")
                detections_json = {
                    "filename": filename,
                    "detections": detections_list
                }
                with open(json_file_path, 'w') as f:
                    json.dump(detections_json, f, indent=2)
                print(f"üìÑ Saved detections log for {filename} to {json_file_path}")

            except Exception as e:
                print(f"‚ùå Error processing image {filename}: {e}")
```
